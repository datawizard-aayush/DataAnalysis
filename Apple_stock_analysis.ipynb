```python
# 📊 Apple Stock Data Analysis using yfinance

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import json

# 1. Create Ticker object for Apple
apple = yf.Ticker("AAPL")

# 2. Load Apple company info from JSON
!wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/data/apple.json

with open('apple.json') as json_file:
    apple_info = json.load(json_file)

# 3. Explore info
print("\n \n Type of data:", type(apple_info))
print("\n \n Sector:", apple_info['sector'])

# 4. Get full Apple stock price history
apple_share_price_data = apple.history(period="max")
print("\n\n",apple_share_price_data)

# 5. First day volume traded
first_day_volume = apple_share_price_data.iloc[0]['Volume']
print("\n\n First day Volume traded:", first_day_volume)

# 6. Summary statistics
print("\n\n Descriptive Statistics:\n", apple_share_price_data.describe())

# 7. Reset index to get 'Date' as a column
apple_share_price_data.reset_index(inplace=True)
print("\n\n First 5 rows after resetting index:\n", apple_share_price_data.head())

# 8. Plot stock opening price over time
apple_share_price_data.plot(x="Date", y="Open", title="Apple Share Opening Price")
plt.xlabel("Date")
plt.ylabel("Open Price (USD)")
plt.grid(True)
plt.show()

# 9. View dividend history
print("\n \n Dividend History:\n", apple.dividends)

# 10. Plot dividend history
apple.dividends.plot(title="Apple Dividend History")
plt.xlabel("Date")
plt.ylabel("Dividend (USD)")
plt.grid(True)
plt.show()
```
